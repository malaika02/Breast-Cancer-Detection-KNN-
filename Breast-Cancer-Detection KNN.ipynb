```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df.drop(columns = ['id','Unnamed: 32'],inplace=True)
```

```python
df
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['diagnosis']),df['diagnosis'],test_size=0.2)
```

```python
X_train.shape
```

```python
#in knn model it is good to take data on same scale bcz knn completely rely on calculating distance or agr value bht brri ha 
# to wo dominate kr jye gi so sb columns ko same scale py lye gy
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train) #training wly data pr mean or std count kia fr usko transorm kr dia 
X_test = scaler.transform(X_test) # same mean or std ki values use kry gy
```

```python
X_train
```

```python
X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = knn.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
knn.score(X_test,y_test)
```

```python
#how to select k
#1. Heuristic approach --> square root of n(total inputs) --> avoid even numbers
#2. Experimental Approach --> Cross Validation 

#Cross Validation

scores = []
for i in range(1,16):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred = knn.predict(X_test)
    scores.append(accuracy_score(y_pred,y_test))
```

```python
 scores  #accuracy scores for 15 different models
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,16),scores)
```
